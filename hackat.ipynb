```python
from GPT2LM import gpt2
```

```python
block_size=128 #Example block_size, you can choose your own
text=gpt2.load_text('/kaggle/input/javascripts/train.txt') # Can use load_csv() or load_json() for csv and json 
train_dataset=gpt2.CharDataset(text, block_size, '/kaggle/working/') # Encoder.json will be save in out_dir
```

```python
train_dataset.vocab_size
```

```python
model_config = gpt2.GPT.get_default_config()
model_config.model_type = 'gpt-micro' # more model typesare available, use `print_config()` to see all
model_config.vocab_size = train_dataset.vocab_size
model_config.block_size = block_size  
model = gpt2.GPT(model_config)
```

```python
train_config = gpt2.Trainer.get_default_config()
train_config.learning_rate = 5e-4 
train_config.max_iters = 1000 #iters per epoch
train_config.batch_size = 64
trainer = gpt2.Trainer(train_config, model, train_dataset)
```

```python
trainer.run(15) # number of epochs = 10
```

```python
javascript_code = 'function factorial(n) {\n' + \
                  '    if (n === 0 || n === 1) {\n' + \
                  '        return 1;\n' + \
                  '    } else {\n' + \
                  '        return n * factorial(n - 1);\n' + \
                  '    }\n' + \
                  '}\n' + \
                  '\n' + \
                  'console.log("Factorial of 5:", factorial(5));'
```

```python
print(gpt2.Generate(model, javascript_code, '/kaggle/working/encoder.json', 400))  # 'Hi' is prompt and 200 is the maximum length of generated response means 200 characters will be generated
```

```python
gpt2.save('/kaggle/working/javascript.pth', model)
```

```python
encoder_path = "/kaggle/working/encoder.json"
fine_data=gpt2.load_text('/kaggle/input/javascripts/fine.txt')
fine_config = gpt2.Trainer.get_default_config()
fine_config.learning_rate = 5e-4
fine_config.max_iters = 500
fine_config.batch_size = 64
finetuner = gpt2.FineTuner(model, fine_data, encoder_path, block_size, fine_config)
```

```python
finetuner.fine_tune(5)
```
